```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/MaxVieiraSantiago/Graphics/master/Dataset/chicken.csv'

dados = pd.read_csv( url )
dados.shape
```

```python
dados.head()
```

```python
dados_grupo = dados.groupby( ['feed' ] )['weight'].sum()
dados_grupo
```

```python
alimentos = list( set( dados.feed ) )

plt.figure( figsize=( 18, 9 ) )
linhas = 2
colunas = 3
posicao = 1

for i in range( len( alimentos ) ):
  plt.subplot( linhas, colunas, posicao )
  sns.histplot( dados.loc[ dados['feed'] == alimentos[ i ] ].weight, kde=True ).set_title( alimentos[ i ] )

  posicao+=1

plt.tight_layout()
```
